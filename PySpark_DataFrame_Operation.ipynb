## **Filtering Operation**

```python
# Extract staff records where compensation exceeds 50,000
print("\n High-Earning Personnel (Salary > 50,000):")
df.filter(df.salary > 50000).show()
```

```python
# Identify team members older than 30
print("\n Senior Staff Profiles (Age > 30):")
df.filter(df.age > 30).show()
```

```python
# Calculate the total workforce count
total_staff = df.count()
print(f"\n Total Employee Population: {total_staff}")

```

```python
# Retrieve records for specific regional hubs (e.g., Delhi)
print("\n Regional Report: Delhi-based Employees:")
df.filter(df.city == "Delhi").show()
```

```python
# identity and compensation data
print("\n Summary : Name and Salary Columns:")
df.select("name", "salary").show()
```

## **GroupBy Operation**

```python
from pyspark.sql import functions as F
```

```python
# Calculate the total workforce distribution by department

print("\n Departmental Breakdown (Employee Counts):")
df.groupBy("department").count().orderBy("count", ascending=False).show()
```

```python
# Compute the average compensation metrics across different departments

print("\n Average Salary by Department:")
df.groupBy("department").agg(F.round(F.avg("salary"), 2).alias("avg_salary")).show()
```

```python
# Analyze geographic density of personnel

print("\n Staff by City:")
df.groupBy("city").count().orderBy("count", ascending=False).show()
```

```python
# Identify the highest salary offering in each professional role

print("\n Highest Salary per Role:")
df.groupBy("role").max("salary").withColumnRenamed("max(salary)", "top_pay").show()
```

```python
# Demographic summary by gender (Average Age and Count)

print("\n Age and Volume by Gender:")
df.groupBy("gender").agg(
    F.round(F.avg("age"), 1).alias("mean_age"),
    F.count("*").alias("total_count")
).show()
```

## **Multiple Aggregations**

```python
# Conduct a global salary audit (Average, Minimum, and Maximum)

print("\n Enterprise Salary Benchmarking (Global Metrics):")
df.select(
    F.avg("salary").alias("average_compensation"),
    F.min("salary").alias("entry_level_pay"),
    F.max("salary").alias("executive_level_pay")
).show()
```

```python
# Perform a detailed departmental performance and cost analysis

print("\n Staff Count, Avg Age, & Total Budget:")
df.groupBy("department").agg(
    F.count("employee_id").alias("headcount"),
    F.round(F.avg("age"), 1).alias("mean_age"),
    F.sum("salary").alias("total_payroll_expenditure")
).orderBy("total_payroll_expenditure", ascending=False).show()
```

```python
# Analyze salary variance within professional roles

print("\n Pay Range Min vs Max per Role:")
df.groupBy("role").agg(
    F.min("salary").alias("floor_salary"),
    F.max("salary").alias("ceiling_salary"),
    (F.max("salary") - F.min("salary")).alias("pay_gap")
).show()
```

```python
# Evaluate geographic operational costs and workforce seniority

print("Total Staff & Average Seniority:")
df.groupBy("city").agg(
    F.count("*").alias("regional_headcount"),
    F.round(F.avg("salary"), 0).alias("avg_regional_pay")
).orderBy("avg_regional_pay", ascending=False).show()
```
