# **Pyspark Dataframe Basics**


### **1. importing pyspark**

```python
import pyspark
from pyspark.sql import SparkSession
```

### **2. start spark session**

```python
sparks = SparkSession.builder.appName('Dataframe').getOrCreate()
```

### **3. connect to drive**

```python
from google.colab import drive
drive.mount('/content/drive')
```

### **4. read the dataset**

```python
df = sparks.read.option("header", "true").csv("/content/drive/MyDrive/dataset/IMDB Dataset.xls")
```

```python
df.show(5)
```

### **5. Check the datatype of the dataframe**

```python
df.printSchema()
```

### **6. check the summary**

```python
df.describe().show()
```

### **7. select column**

```python
df.select(['review']).show()
```

### **8. Adding column**

```python
from pyspark.sql import functions as F
df = df.withColumn('review_length', F.length(df['review']))
```

```python
df.show(5)
```

### **9. Dropping column**

```python
df = df.drop('review_length')
```

```python
df.show(5)
```

### **10.Renaming column**

```python
df = df.withColumnRenamed('sentiment', 'emotion')
```

```python
df.show(5)
```
